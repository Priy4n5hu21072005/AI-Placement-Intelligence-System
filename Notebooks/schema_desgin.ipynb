```python
final_columns = [
    "cgpa",
    "college_tier",
    "branch",
    "internship_count",
    "project_count",
    "skills_score",
    "communication_score",
    "certifications",
    "dsa_score",
    "placed"
]
import os
import pandas as pd

# Go one level up (from Notebooks â†’ Major_Project)
BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), ".."))

# Raw data folder
DATA_DIR = os.path.join(BASE_DIR, "Data", "Raw")

# File paths
placement_train_path = os.path.join(DATA_DIR, "Student_placement_train.csv")
placement_test_path = os.path.join(DATA_DIR, "Student_placement_test.csv")
student_perf_path = os.path.join(DATA_DIR, "StudentPerformanceFactors.csv")
stack_path = os.path.join(DATA_DIR, "stackoverflow_full.csv")

# Load
train_df = pd.read_csv(placement_train_path)
test_df = pd.read_csv(placement_test_path)
perf_df = pd.read_csv(student_perf_path)
stack_df = pd.read_csv(stack_path)
 # merge train and test 
placement_df=pd.concat([train_df,test_df],axis=0)
placement_df.reset_index(drop=True,inplace=True)
print(placement_df.shape)
placement_df.head()
placement_df.columns
```

```python
final_df=pd.DataFrame()
final_df['cgpa']=placement_df['CGPA']
final_df['college_tier']=2
final_df['branch']=placement_df['Branch']
final_df['internship_count']=placement_df['Internships']
final_df['project_count']=placement_df['Projects']
final_df['skills_score']=placement_df['Coding_Skills']
final_df["communication_score"]=placement_df['Communication_Skills']
final_df["certifications"]=placement_df['Certifications']
final_df['dsa_score']=placement_df['Coding_Skills']
final_df['placed']=placement_df['Placement_Status']
```

```python
# covert placed =1 or 0
final_df["placed"] = placement_df["Placement_Status"].str.strip()

final_df["placed"] = final_df["placed"].map({
    "Placed": 1,
    "Not Placed": 0
})

final_df['placed'].value_counts()
final_df.isnull().sum()
```

```python
processed_path = os.path.join(BASE_DIR, "Data", "processed", "placement_master.csv")
final_df.to_csv(processed_path, index=False)
```
